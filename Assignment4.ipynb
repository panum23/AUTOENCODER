```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('creditcard.csv')
data
```

```python
data=data.drop(['Time','Class'],axis=1)
data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
xtrain,xtest=train_test_split(data,test_size=0.2)
```

```python
print(xtrain.shape[1])
print(xtest.shape[1])
type(xtrain)
```

```python
from keras.models import Model,Sequential
from tensorflow.keras.layers import Dense,Dropout
from tensorflow.keras import layers,models
```

```python
encoder=tf.keras.models.Sequential([
    layers.Input(shape=(xtrain.shape[1],)),
    layers.Dense(256,activation='relu'),
    layers.Dense(128,activation='relu'),
    layers.Dense(64,activation='relu'),
    layers.Dense(32,activation='relu'),
])

decoder=tf.keras.models.Sequential([
    layers.Input(shape=(32,)),
    layers.Dense(64,activation='relu'),
    layers.Dense(128,activation='relu'),
    layers.Dense(256,activation='relu'),
    layers.Dense(xtrain.shape[1],activation='linear')
])

model= tf.keras.models.Sequential([
    encoder,
    decoder
])
```

```python
model.compile(optimizer='adam',loss='mse')
```

```python
history=model.fit(
    xtrain,
    xtrain,
    validation_data=(xtest,xtest),
    epochs=10,
    batch_size=100,
    shuffle=True
)
```

```python
import seaborn as sea
sea.lineplot(model.history.history)
```

```python
predictions =model.predict(xtest)
mse=np.mean(np.power(xtest-predictions,2),axis=1)
```

```python
threshold=np.percentile(mse,95)
threshold
```

```python
anomalies=mse>threshold
```

```python
num_anomilies=np.sum(anomalies)
print(num_anomilies)
```

```python
import matplotlib.pyplot as plt
plt.plot(mse,marker='o',linestyle='',markersize=3,label='MSE')
plt.axhline(threshold,color='r',linestyle='--',label='Anomaly Threshold')
plt.xlabel('Sample Index')
plt.ylabel('MSE')
plt.title('Anomaly Detection Results')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(xtest.iloc[0], label='Original ECG')
plt.plot(predictions[0], label='Reconstructed ECG')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.title('Normal ECG')
plt.show()
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix, classification_report

plt.figure(figsize = (6, 4.75))
sns.heatmap(confusion_matrix(anomalies, anomalies), annot = True, annot_kws = {"size": 16}, fmt = 'd')

plt.xlabel("Predicted label", fontsize = 14)
plt.ylabel("True label", fontsize = 14)
plt.title("Confusion Matrix", fontsize = 14)
plt.grid(False)
plt.show()
```
